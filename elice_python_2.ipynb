# Quiz 로또번호 생성

```python
import numpy as np

def make_lotto(count):
    for i in range(count):
        lotto_num = []
        for j in range(6):
            lotto_num = numpy.random.choice(range(1, 46), 6, replace = False)
            # 복원 추출을 막아주세요 == replace (중복 방지)
            # nCr : n 개 중에 r 을 순서와 상관없이 뽑이주세요.
            lotto_num.sort()
        print('{} 로또번호 : {}'.format(i+1, lotto_num))
```

```python
count = int(input("로또 번호를 몇 개 생성할까요?"))
make_lotto(count)
```
